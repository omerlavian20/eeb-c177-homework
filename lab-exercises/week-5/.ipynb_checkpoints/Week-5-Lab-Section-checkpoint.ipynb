```python
import matplotlib.pyplot as plt
```

```python
#heatmap = visualizing data with color intensity-->low res image
#data has patients on y axis, days on x

image = plt.imshow(data)
plt.show()
#requirement for command to work in Jupyter
```

```python
#x=# of days, y=patients, intensity=inflammation score
#see from this that data goes up then down over time
```

```python
avg_inflammation = np.mean(data, axis=0)
#rows are axis 0
#taking average for all patients for each day
avg_plot=plt.plot(avg_inflammation)
plt.xlabel('time(days)')
plt.ylabel('inflammation score')
plt.title('patient data')
plt.show()
#similar to heat map
#ONLY DO 1 PLT.SHOW SO CAN actuallly label axes
```

```python
max_plot = plt.plot(np.max(data, axis=0))
plt.show()
#more sharp peaks than above because just maxes
```

```python
min_plot = plt.plot(np.min(data, axis=0))
plt.show()
```

```python
#step 1 is to make a figure object, which can hold subfigures/mulitple plots
fig = plt.figure(figsize=(10.0,3.0))

sbplt1 = fig.add_subplot(1,3,1)
sbplt2 = fig.add_subplot(1,3,2)
sbplt3 = fig.add_subplot(1,3,3)
#first two numbers gives dimensions of figure 
#third number is position number

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0), '--')

sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data,axis=0),'o-')

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))
#sbplt makes things one image


fig.tight_layout()
#keeps everything all together at same size

plt.show()

#homework involves same thing but making plots vertical; consider switching figure size AND FIGURE DIMENSIONS
```

```python
#unlike other languages, don't need to make pointer for memory address
#if set one variable equal to another variable, have same info because reference same location in memory (shallow copy)-->not making a copy really-->if edit new variable old one changes as well
#deep copy you set equal to list(original list); makes computer take old list and move information from old list into a new list blocked in different spot in computer

salsa = ['peppers', 'onions', 'cilantro', 'tomatoes']
print('salsa1:', salsa)
my_salsa=list(salsa)
#shallow copy before
#list makes deep copy
print('salsa2', my_salsa)
```

```python
print(enumerate(salsa))
print(enumerate(my_salsa))
#same memory address-->shallow copy
```

```python
mysalsaa = deepcopy(salsa)
```

```python
print(enumerate(mysalsaa))
```

```python
salsa = ['peppers', 'onions', 'cilantro', 'tomatoes']
my_salsa = salsa
salsa[0] = 'hot peppers'
print('ingredients in my salsa:', my_salsa)
#make copy with list function, i.e. deep copy, if want to change one copy of list without changing og
```

```python
#get all even nums from 1 to 100
evens = []
for ii in range(1,101):
    if (ii % 2 == 0):
        evens.append(ii)
        
print(evens)

#get odds
odds = []
for ii in range (1,101):
    if ii % 2 == 1:
        odds.append(ii)
print('\n')
print(odds)
```

```python
#find squares for first ten even numbers
squares_nums = [i*i for i in range(20) if i % 2 == 0]
print(squares_nums)
```
