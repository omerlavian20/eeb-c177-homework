## Part 1 (Figures)

```python
import numpy as np
data = np.loadtxt(fname='inflammation-01.csv', delimiter=",")
import matplotlib.pyplot as plt
```

1.)

```python
#note to self: DON'T FORGET plt.show()!!!!!! kind of like how need to type "print()" for certain things
#inflammtion data format: x-axis (axis=0) is inflammation data recorded on each of the days, y-axis (axis = 1) is 
#such that each row represents one patient-->each column represents one day
standard=np.std(data, axis=0)
stand_plot=plt.plot(standard)
plt.xlabel('time (days)')
plt.ylabel('standard deviation')
plt.title("Standard Deviation in Inflammation in All Patients Across Whole Timeline")
plt.show()
```

2.) I'm assuming this is referring to the three plots we made in lab, so I just pasted that code and made adjustments.

```python
#step 1 is to make a figure object, which can hold subfigures/mulitple plots
fig = plt.figure(figsize=(3.0,10.0))

sbplt1 = fig.add_subplot(3,1,1)
sbplt2 = fig.add_subplot(3,1,2)
sbplt3 = fig.add_subplot(3,1,3)
#first two numbers gives dimensions of figure 
#third number is position number

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0), '--')

sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data,axis=0),'o-')

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))
#sbplt makes things one image

plt.xlabel('time(days)')

fig.tight_layout()
#keeps everything all together at same size

plt.show()
```
